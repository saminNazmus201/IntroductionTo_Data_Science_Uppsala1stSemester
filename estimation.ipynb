```python
def linear_regression(x,y):
    # Here x,y is our data
    from sklearn.linear_model import LinearRegression
    lr = LinearRegression()
    lr.fit(x,y)
    return lambda x1: lr.predict(x1)
```

```python
import random
import numpy as np
def sample_data(n):
    X= np.random.uniform(0,1, size=n).reshape(-1,1)
    Y= 3*X.flatten()+np.random.normal(0,1, size= len(X[0]))
    return X, Y
```

```python
sample_data(10)
```

```python
x_train, y_train= sample_data(10)
ghat= linear_regression(x_train, y_train)
```

```python
def test_error():
    xtest,ytest=sample_data(100)
    y_pred= ghat(xtest)
    error= ytest-y_pred
    return np.mean(error**2)
```

```python
import matplotlib.pyplot as plt
plt.hist([test_error() for i in range(100)])
```

```python
from sklearn.datasets import load_iris
data= load_iris()
X= data.data
Y= data.target
X_train1= X[:100, :]
Y_train1= Y[:100]
X_test1= X[100:, :]
Y_test1 = Y[100:]
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(X_train1, Y_train1)
```

```python
pred= lr.predict(X_test1)
loss= np.mean(Y_test1!=pred)
loss
```
